## ElasticNet (alpha=1)

```python
# train, test 분리
X_train, X_test, y_train, y_test  = train_test_split(df.iloc[:,:-1], df["score_절폭"],test_size = 0.3, random_state=0)
```

```python
elastic = ElasticNet(alpha=1)
elastic.fit(X_train, y_train)
mean_absolute_error(elastic.predict(X_test), y_test)
```
