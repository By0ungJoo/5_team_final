```python
import numpy as np
import pandas as pd
```

```python
gicho_2017 = pd.read_csv('2017.csv')
gicho_2018 = pd.read_csv('2018.csv')
gicho_2019 = pd.read_csv('2019.csv')
gicho_2020 = pd.read_csv('2019.csv')
```

```python
gicho_2017.drop('Unnamed: 0', axis = 1, inplace = True)
gicho_2018.drop('Unnamed: 0', axis = 1, inplace = True)
gicho_2019.drop('Unnamed: 0', axis = 1, inplace = True)
gicho_2020.drop('Unnamed: 0', axis = 1, inplace = True)
gicho_2017 = gicho_2017.rename(columns={'index' : 'police'})
gicho_2018 = gicho_2018.rename(columns={'Unnamed: 0.1' : 'police'})
gicho_2019 = gicho_2019.rename(columns={'Unnamed: 0.1' : 'police'})
gicho_2020 = gicho_2020.rename(columns={'Unnamed: 0.1' : 'police'})
gicho_2017['year'] = 2017
gicho_2018['year'] = 2018
gicho_2019['year'] = 2019
gicho_2020['year'] = 2020
```

```python
gicho_2017.groupby(['year', 'police']).sum()
```

```python
gicho_2017 = gicho_2017.groupby(['year', 'police'], sort = True).sum()
gicho_2018 = gicho_2018.groupby(['year', 'police'], sort = True).sum()
gicho_2019 = gicho_2019.groupby(['year', 'police'], sort = True).sum()
gicho_2020 = gicho_2020.groupby(['year', 'police'], sort = True).sum()
```

```python
df = pd.concat([gicho_2017, gicho_2018, gicho_2019, gicho_2020], axis = 0)
```

```python
df['총합'] = df.apply(np.sum, axis =1)
```

```python
df.to_csv('기초수급자현황.csv')
```
