## 2. 관할서 지정

```python
df_bl
```

```python
from tqdm import tqdm
```

```python
# 공원 위도경도가 관할경계에 들어가는지 파악하고, 들어간다면 관할서 지정해주는 코드
for i in tqdm(range(41)):
    for z in range(18334):
        if df_bl["geometry"][z].within(df_js["geometry"][i]):
            df_bl.loc[z,"관할서"]=df_js["NAME"][i]
            
```

## 3.전처리

```python
import pandas as pd
```

```python
df = pd.read_csv("데이터별_관할서리스트/비상벨현황_관할서리스트.csv", index_col = 0)
df
```

```python
# 관할서명  수정
df["관할서"] = df["관할서"].str.replace("마산", "경남마산")                                     
df["관할서"] = df["관할서"].str.replace("창원", "경남창원")
df["관할서"] = df["관할서"].str.replace("수원", "경기수원")
df["관할서"] = df["관할서"].str.replace("진주", "경남진주")
df["관할서"] = df["관할서"].str.replace("진해", "경남진해")
df["관할서"] = df["관할서"].str.replace("세종", "충남세종")                                           
```

```python
#관할서별 비상벨 개수 데이터프레임화
df_bell = pd.DataFrame(df.groupby("관할서", sort = True).count()["bell_us"]).reset_index()
df_bell.rename(columns={"bell_us":"bell_cnt"}, inplace=True)
df_bell
```

```python
##도봉경찰서가 없네...
### 도봉구 = 622개 있대...
df_bell.loc[40,:]=["서울도봉경찰서",622]
df_bell
```

```python
#파일저장
df_bell.sort_values("관할서").to_csv("전처리완성본/22_관할서별_비상벨개수.csv")
```
