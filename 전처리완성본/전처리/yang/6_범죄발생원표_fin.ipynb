```python
import pandas as pd
```

```python
df = pd.read_csv("raw_data/6.범죄발생원표.csv")
df.head()
```

```python
##17년도 이전 데이터가 있는 것 확인
##17,18,19 년도만 추출
df = df[df["crm_date"].astype(str).str.contains("2017|2018|2019")]
df.head()
```

```python
df.info()
```

## 1. 날짜데이터 통일

```python
#편하게 보기 위해 str변환
date = df["crm_date"].astype(str)
date
```

```python
##"/" 없는 값에 "/"추가 한후, 변경
z = []
for i in date[-date.str.contains("/")]:
    b = list(i)
    b.insert(4,"/")
    b.insert(7,"/")
    a = "".join(b)
    z.append(a)
    
date[-date.str.contains("/")] = z
```

```python
#date값 "/"로 통일
date = date[date.str.contains("/")]
date
```

```python
#df의 날짜데이터 변경 후 다시 str로 변경(분석시 편함)
df["crm_date"] = pd.to_datetime(date)

#index 재조정
df.reset_index(drop=True, inplace = True)
df.tail()
```

```python
#df["crm_date"][0].strftime("%Y%m%d")

num = len(df)
z = []
for i in range(num):
    a = df["crm_date"][i].strftime("%Y%m%d")
    z.append(a)
    
z
```

```python
df["crm_date"] = z
df.head()
```

```python
df.to_csv("6_범죄발생원표_날짜전처리_완성본")
```

## 2. 연도별 정리

```python
df = pd.read_csv("6_범죄발생원표_날짜전처리_완성본")
```

```python
df["crm_date"] = df["crm_date"].astype(str)
```

```python
#연도별 정리
df_2017 = df[df["crm_date"].str.contains("2017")].iloc[:,1:]
df_2018 = df[df["crm_date"].str.contains("2018")].iloc[:,1:]
df_2019 = df[df["crm_date"].str.contains("2019")].iloc[:,1:]
```

---

## 3. 연도별, 관할서별 절도 폭력 발생건수

```python
df_2017.head()
```

```python
#관할서 공백으로 인한 전처리 필요성 확인
df_2017["jur_stn"] = df_2017["jur_stn"].str.strip()
df_2018["jur_stn"] = df_2018["jur_stn"].str.strip()
df_2019["jur_stn"] = df_2019["jur_stn"].str.strip()
```

```python
# 연도별 , 관할서별 절도,폭력  발생 건수(관할서  가나다순 정리)
gr_2017_절폭 = df_2017[df_2017["5m_crm_yn"].isin(["절도","폭력"])].groupby("jur_stn",sort = True).count()["crm"]
gr_2018_절폭 = df_2018[df_2018["5m_crm_yn"].isin(["절도","폭력"])].groupby("jur_stn",sort = True).count()["crm"]
gr_2019_절폭 = df_2019[df_2019["5m_crm_yn"].isin(["절도","폭력"])].groupby("jur_stn",sort = True).count()["crm"]
```

```python
#관할서, 연도별_절도폭력 데이터 프레임 생성
df_절폭 = pd.DataFrame({"관할서":gr_2017_절폭.index, "발생_17_절도폭력":gr_2017_절폭.values, "발생_18_절도폭력":gr_2018_절폭.values, 
              "발생_19_절도폭력":gr_2019_절폭.values})
df_절폭.head()
```

## 4. 연도별, 관할서별 강도 살인 발생건수

```python
# 연도별 , 관할서별 절도,폭력  발생 건수(관할서  가나다순 정리)
gr_2017_강살 = df_2017[df_2017["5m_crm_yn"].isin(["강도","살인"])].groupby("jur_stn",sort = True).count()["crm"]
gr_2018_강살 = df_2018[df_2018["5m_crm_yn"].isin(["강도","살인"])].groupby("jur_stn",sort = True).count()["crm"]
gr_2019_강살 = df_2019[df_2019["5m_crm_yn"].isin(["강도","살인"])].groupby("jur_stn",sort = True).count()["crm"]
```

```python
#관할서, 연도별_절도폭력 데이터 프레임 생성
df_강살 = pd.DataFrame({"관할서":gr_2017_강살.index, "발생_17_강도살인":gr_2017_강살.values, "발생_18_강도살인":gr_2018_강살.values, 
              "발생_19_강도살인":gr_2019_강살.values})
df_강살.head()
```

## 5. 연도별, 관할서별 교통사고 발생건수

```python
## crm 에서 교통, 도로, 사고, 도주차량의 단어가 포함된 항목 추출
gr_2017_교통 = df_2017[df_2017["crm"].str.contains("교통|도로|사고|도주차량")].groupby("jur_stn",sort = True).count()["crm"]
gr_2018_교통 = df_2018[df_2018["crm"].str.contains("교통|도로|사고|도주차량")].groupby("jur_stn",sort = True).count()["crm"]
gr_2019_교통 = df_2019[df_2019["crm"].str.contains("교통|도로|사고|도주차량")].groupby("jur_stn",sort = True).count()["crm"]
```

```python
#관할서, 연도별_절도폭력 데이터 프레임 생성
df_교통 = pd.DataFrame({"관할서":gr_2017_교통.index, "발생_17_교통사고":gr_2017_교통.values, "발생_18_교통사고":gr_2018_교통.values, 
              "발생_19_교통사고":gr_2019_교통.values})
df_교통.head()
```

## 6. 법질서준수도( 기초질서, 집회지위질서 등 -> 5대범죄, 교통사고 제외)


```python
##연도별 , 관할서별, 기초질서 = 5대범죄 , 교통사고 제외 하고 나머지
gr_2017_질서 = df_2017[df_2017["5m_crm_yn"]==' ']
gr_2017_질서 = gr_2017_질서[-df_2017_질서["crm"].str.contains("교통|도로|사고|도주차량")].groupby("jur_stn",sort = True).count()["crm"]

gr_2018_질서 = df_2018[df_2018["5m_crm_yn"]==' ']
gr_2018_질서 = gr_2018_질서[-df_2018_질서["crm"].str.contains("교통|도로|사고|도주차량")].groupby("jur_stn",sort = True).count()["crm"]

gr_2019_질서 = df_2019[df_2019["5m_crm_yn"]==' ']
gr_2019_질서 = gr_2019_질서[-df_2019_질서["crm"].str.contains("교통|도로|사고|도주차량")].groupby("jur_stn",sort = True).count()["crm"]
```

```python
#관할서, 연도별_절도폭력 데이터 프레임 생성
df_질서 = pd.DataFrame({"관할서":gr_2017_질서.index, "발생_17_기초질서":gr_2017_질서.values, "발생_18_기초질서":gr_2018_질서.values, 
              "발생_19_기초질서":gr_2019_질서.values})
df_질서.head()
```

## 7. 전반적 안전도(ALL, 모든범죄건수)

```python
# 전반적 안전도 = 관할서별 범죄건수
gr_2017_all = df_2017.groupby("jur_stn", sort = True).count()["crm"]
gr_2018_all = df_2018.groupby("jur_stn", sort = True).count()["crm"]
gr_2019_all = df_2019.groupby("jur_stn", sort = True).count()["crm"]
```

```python
#관할서, 전반적 안전도 데이터 프레임 생성
df_all = pd.DataFrame({"관할서":gr_2017_all.index, "발생_17_전반":gr_2017_all.values, "발생_18_전반":gr_2018_all.values, 
              "발생_19_전반":gr_2019_all.values})
df_all.head()
```

----
## 8. 모든파일 병합

```python
df_범죄발생원표 = pd.concat([df_절폭, df_강살.iloc[:,1:] , df_교통.iloc[:,1:], df_질서.iloc[:,1:] , df_all.iloc[:,1:]], axis = 1)
df_범죄발생원표.head()
```

## 9. 전처리 완성본 파일 저장

```python
df_범죄발생원표.to_csv("6_범죄발생원표_전처리완성본")
```
